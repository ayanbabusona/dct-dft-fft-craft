```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['DCT', 'DWT', 'FFT']
students = [img3_text_dct_std,img3_text_dwt_std,img3_text_fft_std]
ax.bar(langs,students)
plt.xlabel("Image Enhancement Technique")
plt.ylabel("Standard Deviation")
fig.suptitle("For text region", fontsize = 16)
plt.show()
```
