# DCT

```python
src_img = cv2.imread('img_4.jpeg', 0)
```

```python
src_img_1 = src_img.astype('float')
```

```python
img_dct = cv2.dct(src_img_1)
```

```python
img_reconst = cv2.idct(img_dct)
```

```python
plt.imshow(img_reconst)
plt.axis('off')
```

```python
cv2.imwrite('1_dct.jpg', img_reconst)
```

# DWT

```python
coeffs = pywt.dwt2(img_reconst, 'db3', mode = 'periodization')
```

```python
cA, (cH, cV, cD) = coeffs # Extracting coefficients
```

```python
imgr = pywt.idwt2(coeffs, 'db3', mode = 'periodization')
```

```python
imgr = np.uint8(imgr)
```

```python
plt.imshow(imgr)
plt.axis('off')
```

```python
cv2.imwrite('2_dwt.jpg', imgr)
```

# FFT

```python
fft_output = np.fft.fft2(imgr)
```

```python
fft_reconst = np.fft.ifft2(fft_output)
```

```python
plt.imshow(np.abs(fft_reconst))
plt.axis('off')
```

```python
cv2.imwrite('3_fft.jpg', np.abs(fft_reconst))
```
